### Gini Impurity

#### Ruas Kiri:
$G = 1 - \displaystyle\sum_{i}^nP^2_{i}$
    
$G = 1 - P(biru)^2$

$G = 1 - (\frac{4}{4})^2 = 0$
#### Ruas Kanan:
$G = 1 - \displaystyle\sum_{i}^nP^2_{i}$

$G = 1 - (P(biru)^2+P(hijau)^2)$

$G = 1 - ((\frac{1}{6})^2+(\frac{5}{6})^2) = 0.278$
#### Average Gini Impurity:
$G = \frac{4}{4+6}\times 0 + \frac{6}{4+6}\times 0.278$

$G = 0.1668$
